```python
# btvn 1
# cho ma tran [[99, 99, 99],
#              [99, 99, 99]
#              [99, 99, 99]]

# hay tuong tuong o 99 trong
# gia su 1 la x
# gia su 0 la 0

# lay dau vao tu phia x va phia o luan phien
# neu dau vao ti phia x la (0,0)

# thi ma tra se la
# [[x, 99, 99],
#  [99, 99, 99],
#  [99, 99, 99]]

# neu dau vao tu phi 0 cung la (0,0) thi thong bao ho nhap lai
# neu khong thi dien vao ma tran

# thu thach: neu nguoi dau tien co 3 phan tu lien tiep thi dung tro choi
```

```python
import numpy as np

# hàm nhập vị trí người chơi A
def a_input():
    xa,ya=list(input("\nNhap vi tri nguoi choi A: ").split(','))
    return xa,ya

# hàm nhập vị trí người chơi B
def b_input():
    xb,yb=list(input("\nNhap vi tri nguoi choi B: ").split(','))
    return xb,yb

# hàm kiểm tra vị trí người chơi A có hợp lệ hay không, nếu không hợp lệ thì yêu cầu nhập lại
def a_validate(matrix, xa, ya):
    
    while matrix[int(xa)][int(ya)] != 99:
        
        xa, ya = a_input()
        
        if matrix[int(xa)][int(ya)] == 99:
            break
        
        elif matrix[int(xa)][int(ya)] ==1:
            print("\nVi tri da duoc ban chon, vui long chon vi tri khac")
            continue
        
        else:
            print("\nVi tri da duoc nguoi choi B, vui long chon vi tri khac")
            continue
    
    matrix[int(xa)][int(ya)] = '1'
    print("\n",matrix)
    return matrix, xa, ya

# hàm kiểm tra vị trí người chơi B có hợp lệ hay không, nếu không hợp lệ thì yêu cầu nhập lại
def b_validate(matrix, xb, yb):
    
    while matrix[int(xb)][int(yb)] != 99:
        
        xb, yb = b_input()
        
        if matrix[int(xb)][int(yb)] == 99:
            break
        
        elif matrix[int(xb)][int(yb)] ==0:
            print("\nVi tri da duoc ban chon, vui long chon vi tri khac")
            continue
        
        elif matrix[int(xb)][int(yb)] ==1:
            print("\nVi tri da duoc nguoi choi A, vui long chon vi tri khac")
            continue
        
        else:
            print("\nVi tri khong hop le, vui long chon vi tri khac")
            continue
    
    matrix[int(xb)][int(yb)] = '0'
    print("\n",matrix)
    return matrix, xb, yb

# hàm boolean kiểm tra điều kiện chiến thắng tick tack toe 
def win(matrix):
    
    # kiểm tra các hàng, cột và đường chéo để xác định người chiến thắng
    # nếu tìm ra người chiến thắng thì trả về True, ngược lại trả về False
    
    if (matrix[0, 0:3] == 1).all() or (matrix[1, 0:3] == 1).all() or (matrix[2, 0:3] == 1).all() or (matrix[0:3, 0] == 1).all() or (matrix[0:3, 1] == 1).all() or (matrix[0:3, 2] == 1).all() or (matrix.diagonal() == 1).all() or (np.fliplr(matrix).diagonal() == 1).all():
        print("=== Nguoi choi A da chien thang! ===")
        return True
    
    elif (matrix[0, 0:3] == 0).all() or (matrix[1, 0:3] == 0).all() or (matrix[2, 0:3] == 0).all() or (matrix[0:3, 0] == 0).all() or (matrix[0:3, 1] == 0).all() or (matrix[0:3, 2] == 0).all() or (matrix.diagonal() == 0).all() or (np.fliplr(matrix).diagonal() == 0).all():
        print("=== Nguoi choi B da chien thang! ===")
        return True
    
    else: 
        return False

matrix = np.array([[99,99,99],
                   [99,99,99],
                   [99,99,99]]) 

# vòng lặp chính của trò chơi, nếu chưa tìm ra được người chiến thắng thì trò chơi sẽ tiếp tục
while win(matrix) == False:
        
    xa,ya = a_input()
    a_validate(matrix, xa, ya)

    xb,yb = b_input()
    b_validate(matrix, xb, yb)


        
        
```

```python
from sklearn.model_selection import train_test_split
```

```python
# tao mot mang numpy co kich thuoc 150x5, gia su mang nay chua 150 mau du lieu cua sv bao gom chieu cao, can nang, tuoi, luong va diem trung binh
np.random.seed(42)
data = np.random.rand(150, 5)
```

```python
# tach mang dua tren 4 cot dau tien thanh mot bien co ten la X va cot cuoi cung thanh bien co ten la y
X = data[:, :4]   
y = data[:, 4]  
print(X.shape)  
print(y.shape) 
```

```python
# tach X thanh 2 bien X_train va X_test, trong do X_test chua 30% du lieu ban dau va 
# tach y thanh 2 bien Y_train va y_test, trong do Y_test chua 30% du lieu ban dau
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.3,
    random_state=42
)

print("X_train:", X_train.shape)
print("X_test :", X_test.shape)
```

```python
# tao 10 tap du lieu X_train khong chong cheo len nhau
X_train_splits = np.array_split(X_train, 10)

# kiểm tra kích thước từng tập
for i, split in enumerate(X_train_splits):
    print(f"Tap {i+1}: {split.shape}")
```
